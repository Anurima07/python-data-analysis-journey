```python
import numpy as np

data = np.genfromtxt(
    r"D:\Retail-Sales-Analytics\sales.csv",
    delimiter=",",
    names=True,
    dtype=None,
    encoding="utf-8"
)

print(data[:5])
```

```python
print("Rows:", len(data))
print("Columns:", len(data.dtype.names))
print("Shape:", data.shape)
```

```python
print(data.dtype.names)
```

```python
print(data[:10])
```

```python
sales = data["Sales"]
profit = data["Profit"]
quantity = data["Quantity"]
discount = data["Discount"]
region = data["Region"]
category = data["Category"]
```

```python
print("Total Sales :", np.sum(sales))
print("Total Profit :", np.sum(profit))

print("Average Sales :", np.mean(sales))
print("Average Profit :", np.mean(profit))

print("Maximum Sale :", np.max(sales))
print("Minimum Sale :", np.min(sales))

print("Maximum Profit :", np.max(profit))
print("Minimum Profit :", np.min(profit))

print("Average Discount :", np.mean(discount))

print("Total Quantity :", np.sum(quantity))
```

```python
#Part 1: Region-wise Sales
regions= np.unique(region)
for r in regions:
    total=np.sum(sales[region == r])
    print(r,":",total)
```

```python
#Region_wise profit
for r in regions:
    total=np.sum(profit[region == r])
    print(r,":",total)
```

```python
#Part 3: Category-wise Sales
categories = np.unique(category)

for c in categories:
    total = np.sum(sales[category == c])
    print(c, ":", total)
```

```python
#Part 4: Category-wise Profit
for c in categories:
    total = np.sum(profit[category == c])
    print(c, ":", total)
```

```python
#Part 5: Best Performing Region
region_sales=[]
for r in regions:
    region_sales.append(np.sum(sales[region == r]))
best= np.argmax(region_sales)
print("Best Region:", regions[best])
```

```python
#Part 6: Best Performing Category
catagory_sales = []

for c in categories:
    category_sales.append(np.sum(sales[category == c]))
best = np.argmax(category_sales) 
print("Best Catagories:", categories[best])
```

```python
#Region with highest profit
region_profit=[]

for r in regions:
    region_profit.append(np.sum(profit[region == r]))
best= np.argmax(region_profit)
print("Best Region:", regions[best])
```

```python
#Category with highest profit
category_profit= []

for c in categories:
    category_profit.append(np.sum(profit[category == c]))

best = np.argmax(category_profit)

print("Best Category:", categories[best])
```

```python
#Average sales per region
for r in regions:
    avarage =np.mean(sales[region == r])
    print(r,":",avarage )
```

```python
#Average profit per category
for c in categories:
    avarage = np.mean(profit[category == c])
    print(c, ":", avarage)
```

```python
#Executive KPI Dashboard (NumPy)
#1. Profit Margin (%)
profit_margin = (np.sum(profit) / np.sum(sales)) * 100

print(f"Profit Margin: {profit_margin:.2f}%")
```

```python
#2. Sales Contribution (%) by Region
total_sales = np.sum(sales)

for r in regions:
    region_sales = np.sum(sales[region == r])
    contribution = (region_sales / total_sales) * 100

    print(f"{r}: {contribution:.2f}%")
```

```python
#Category contributing the highest % of total sales
category_sales = []

for c in categories:
    category_sales.append(np.sum(sales[category == c]))

total_sales = np.sum(sales)

for i in range(len(categories)):
    percent = (category_sales[i] / total_sales) * 100
    print(f"{categories[i]} : {percent:.2f}%")

best = np.argmax(category_sales)

print("\nHighest Sales Contribution:", categories[best])
```

```python
#Region contributing the lowest % of total sales
region_sales = []

for r in regions:
    region_sales.append(np.sum(sales[region == r]))

total_sales = np.sum(sales)

for i in range(len(regions)):
    percent = (region_sales[i] / total_sales) * 100
    print(f"{regions[i]} : {percent:.2f}%")

lowest = np.argmin(region_sales)

print("\nLowest Sales Contribution:", regions[lowest])
```
